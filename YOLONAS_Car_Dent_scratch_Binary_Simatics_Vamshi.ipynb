```python
# Import necessary libraries
import os
import shutil
from sklearn.model_selection import train_test_split

# Define paths to train and test folders for both images and labels
train_images_folder = '/content/drive/MyDrive/Dataset/train/images'
train_labels_folder = '/content/drive/MyDrive/Dataset/train/labels'
test_images_folder = '/content/drive/MyDrive/Dataset/test/images'
test_labels_folder = '/content/drive/MyDrive/Dataset/test/labels'

# Create a 'validate' folder with 'images' and 'labels' subfolders
validate_folder = '/content/drive/MyDrive/Dataset'
validate_images_folder = os.path.join(validate_folder, 'images')
validate_labels_folder = os.path.join(validate_folder, 'labels')
os.makedirs(validate_images_folder, exist_ok=True)
os.makedirs(validate_labels_folder, exist_ok=True)

# Get lists of all image and label files in the train and test folders
train_images = [os.path.join(train_images_folder, img) for img in os.listdir(train_images_folder) if img.endswith('.jpg')]
train_labels = [os.path.join(train_labels_folder, lbl) for lbl in os.listdir(train_labels_folder) if lbl.endswith('.txt')]

# Get lists of all image and label files in the test folders
test_images = [os.path.join(test_images_folder, img) for img in os.listdir(test_images_folder) if img.endswith('.jpg')]
test_labels = [os.path.join(test_labels_folder, lbl) for lbl in os.listdir(test_labels_folder) if lbl.endswith('.txt')]

# Split the train images and labels into train and validation sets
train_images, val_images, train_labels, val_labels = train_test_split(train_images, train_labels, test_size=0.2, random_state=42)

# Move images to the appropriate folders for validation
for img in val_images:
    img_name = os.path.basename(img)
    dest_path = os.path.join(validate_images_folder, img_name)
    shutil.move(img, dest_path)

# Move labels to the appropriate folders for validation
for lbl in val_labels:
    lbl_name = os.path.basename(lbl)
    dest_path = os.path.join(validate_labels_folder, lbl_name)
    shutil.move(lbl, dest_path)

# Print the total number of images in each set
print(f"Total Train Images: {len(train_images)}")
print(f"Total Validation Images: {len(val_images)}")
print(f"Total Test Images: {len(test_images)}")
```

```python
# Import necessary components from the 'super_gradients.training' module
from super_gradients.training.losses import PPYoloELoss
from super_gradients.training.metrics import DetectionMetrics_050
from super_gradients.training.models.detection_models.pp_yolo_e import PPYoloEPostPredictionCallback

# Define parameters for training

train_params = {
    'silent_mode': True,  # Enable silent mode during training (no excessive output)
    'average_best_models': True,  # Average the parameters of the best models during training
    'warmup_mode': 'linear_epoch_step',  # Use linear epoch step for learning rate warm-up
    'warmup_initial_lr': 1e-6,  # Initial learning rate for warm-up
    'lr_warmup_epochs': 3,  # Number of epochs for learning rate warm-up
    'checkpoint_path': checkpoint_path,  # Path to store model checkpoints
    'initial_lr': None,  # Initial learning rate (will be set automatically)
    'lr_mode': 'cosine',  # Learning rate schedule: cosine annealing
    'cosine_final_lr_ratio': 0.1,  # Final learning rate as a fraction of the initial learning rate
    'optimizer': 'Adam',  # Use the Adam optimizer
    'optimizer_params': {'weight_decay': 0.0001},  # Adam optimizer weight decay parameter
    'zero_weight_decay_on_bias_and_bn': True,  # Exclude bias and batch norm from weight decay
    'ema': True,  # Use exponential moving average for model parameters
    'ema_params': {'decay': 0.9, 'decay_type': 'threshold'},  # EMA decay parameters
    'model': 'yolo_nas_l',  # Specify the YOLO-NAS-L model
    'max_epochs': 100,  # Train for a maximum of 100 epochs
    'mixed_precision': True,  # Use mixed precision training for faster computations
    'loss': PPYoloELoss(
        use_static_assigner=False,  # Allow dynamic assignment of ground truth boxes to anchors
        num_classes=len(dataset_params['classes']),  # Specify the number of classes
        reg_max=16  # Limit bounding box regression range
    ),
    'valid_metrics_list': [
        DetectionMetrics_050(
            score_thres=0.1,  # Score threshold for object detection
            top_k_predictions=300,  # Consider top 300 predictions for evaluation
            num_cls=len(dataset_params['classes']),  # Specify the number of classes
            normalize_targets=True,  # Normalize ground truth targets for evaluation
            post_prediction_callback=PPYoloEPostPredictionCallback(
                score_threshold=0.01,  # Score threshold for post-prediction processing
                nms_top_k=1000,  # Consider top 1000 predictions for non-maximum suppression
                max_predictions=300,  # Limit the maximum number of predictions
                nms_threshold=0.7  # Non-maximum suppression threshold
            )
        )
    ],
    'metric_to_watch': 'mAP@0.50'  # Monitor mean Average Precision at IoU 0.50
}
```

#Training the model

```python
# Import necessary libraries
import torch
import torch.optim as optim
import os
from super_gradients.training.losses import PPYoloELoss
from super_gradients.training.metrics import DetectionMetrics_050
from super_gradients.training.models.detection_models.pp_yolo_e import PPYoloEPostPredictionCallback
from super_gradients.training import models, Trainer  # Import the Trainer class

# Assuming model is defined by its name as a string in train_params
model_name = train_params["model"]  # Extract the model name from train_params
model = models.get(model_name, num_classes=len(dataset_params['classes']), pretrained_weights="coco")  # Create the model instance

# Instantiate the optimizer
optimizer_name = train_params["optimizer"]
optimizer_params = train_params["optimizer_params"]
optimizer = getattr(optim, optimizer_name)(model.parameters(), **optimizer_params)  # Create the optimizer

# Define the path to store and load the checkpoint
checkpoint_path = '/content/drive/MyDrive/Dataset/checkpoint_backup/checkpoint_yolo.pth'

# Check if a checkpoint exists and load it if available
if os.path.exists(checkpoint_path):
    checkpoint = torch.load(checkpoint_path)
    model.load_state_dict(checkpoint['model_state_dict'])
    optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
    start_epoch = checkpoint['epoch'] + 1  # Continue from the next epoch
else:
    start_epoch = 0  # Start training from the first epoch

# Set up the Trainer instance for training
trainer = Trainer(experiment_name='my_first_yolonas_run', ckpt_root_dir='/content/drive/MyDrive/Dataset/checkpoint_backup/')

# Update train_params with the dynamic model and optimizer
train_params["model"] = model
train_params["optimizer"] = optimizer

# Training loop
for epoch in range(start_epoch, train_params['max_epochs']):
    # Your existing training logic here (update model, calculate loss, backward, optimizer step, etc.)

    # Save checkpoint at the end of each epoch
    torch.save({
        'epoch': epoch,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
    }, checkpoint_path)
```

```python
# Train the model using the Trainer instance
trainer.train(model=model,               # Pass the model to be trained
              training_params=train_params,  # Pass the training parameters
              train_loader=train_data,       # Pass the training data loader
              valid_loader=val_data)         # Pass the validation data loader
```
